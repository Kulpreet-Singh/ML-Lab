```python
from datetime import datetime, date
def age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
Age = df['BirthDate'].apply(age)
```

```python
df.drop(['BirthDate'], axis = 1, inplace = True)
```

```python
df.insert(1, 'Age', Age)
```

```python
#Binning
import numpy as np
bins = np.linspace(min(df['Age']), max(df['Age']), 4)
group_names = ['Young', 'Senior', 'Super Senior']
df['Age'] = pd.cut(df['Age'], bins, labels =  group_names, include_lowest=True)
```

**Part III: Calculating Proximity /Correlation Analysis of two features**

```python
#Similarity in Simple Matching using simple matching coefficient 
male_and_married = 0
male_and_single = 0
female_and_married = 0
female_and_single = 0

for i in range(len(df)):
  if df['Gender_M'].values[i]==1 and df['MaritalStatus_M'].values[i]==1:
    male_and_married+=1
  elif df['Gender_M'].values[i]==1 and df['MaritalStatus_M'].values[i]==0:
    male_and_single+=1
  elif df['Gender_M'].values[i]==0 and df['MaritalStatus_M'].values[i]==1:
    female_and_married+=1
  else:
    female_and_single+=1

SMC = (male_and_married + female_and_single)/(male_and_married + female_and_single + male_and_single + female_and_married)
SMC
```

```python
#Jaccard Similarity
from sklearn.metrics import jaccard_score
jaccard_score(df['MaritalStatus_M'].values, df['Gender_M'].values)
```

```python
#Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity,cosine_distances
cos_sim=cosine_similarity(df['YearlyIncome'].values.reshape(1,-1),df['AvgMonthSpend'].values.reshape(1,-1))
print (f"Cosine Similarity between A and B:{cos_sim}")
```
