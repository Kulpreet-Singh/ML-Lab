### ***Q2 Pre-processing the corpus***

**Remove punctutations**

```
import string
corpus_without_punct=[]
#normalization contains words without punctuations
table=str.maketrans('','',string.punctuation)
for i in range(len(corpus)):
    corpus_without_punct.append("".join([w.translate(table) for w in corpus[i]]))
print(corpus_without_punct)
# for i in range(len(normalization)):
#     normaliz.append(" ".join([i for i in normalization[i].split() if i.isalpha()]))
# # normaliz
```

**Word Tokenization**

```
corpus_tokenized = [corpus_without_punct[i].split(' ') for i in range(len(corpus_without_punct))]
corpus_tokenized
```

**NORMALIZATION**

```
#Convert to lower case and discard numeric, special chars
normalizedCorpus2 = []
for i in range(len(corpus_tokenized)):
    temp=[]
    temp.append(" ".join([i.lower() for i in corpus_without_punct[i].split() if i.isalpha()]))   
    normalizedCorpus2.append(temp)
normalizedCorpus2
```

**Stopwords Removal**

```
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

```
preProcessedText2 = []
for i in range(len(normalizedCorpus2)):
    temp=[]
    temp.append(" ".join([i for i in normalizedCorpus2[i][0].split() if i not in stop_words]))   #discarding numeric, special chars convert to lower case
    preProcessedText2.append(temp)
preProcessedText2
```

**Stemming**

```
from nltk.stem import PorterStemmer
ps=PorterStemmer()
stemmedText2 = []
for i in range(len(preProcessedText2)):
    temp=[]
    temp.append(" ".join([ps.stem(i) for i in preProcessedText2[i][0].split()]))   #discarding numeric, special chars convert to lower case
    stemmedText2.append(temp)
stemmedText2
```

```
preparedCorpus = []
for i in range(len(stemmedText2)):
  preparedCorpus.append(stemmedText2[i][0])
preparedCorpus
```

### **Q3 a) Bag of Words and tf-idf  (in built)**

```
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import pandas as pd

countvectorizer = CountVectorizer()
tfidfvectorizer = TfidfVectorizer()

count_wm = countvectorizer.fit_transform(preparedCorpus)
tfidf_wm = tfidfvectorizer.fit_transform(preparedCorpus)

count_tokens = countvectorizer.get_feature_names()
tfidf_tokens = tfidfvectorizer.get_feature_names()

df_countvect = pd.DataFrame(data = count_wm.toarray(),index = ['Doc1','Doc2','Doc3','Doc4','Doc5','Doc6','Doc7','Doc8','Doc9','Doc10'],columns = count_tokens)
df_tfidfvect = pd.DataFrame(data = tfidf_wm.toarray(),index = ['Doc1','Doc2','Doc3','Doc4','Doc5','Doc6','Doc7','Doc8','Doc9','Doc10'],columns = tfidf_tokens)

print("Count Vectorizer\n")
print(df_countvect)
print("\nTF-IDF Vectorizer\n")
print(df_tfidfvect)
```

### **Q3 b) Without using in-built functions**

***Bag of Words***

```
#'bag' consists of 10 dictionaries each containing frequency of a word in that particular doc
bag =  []
total_count=[]
for i in range(len(preparedCorpus)):
  c=0
  my_count={} 
  temp = preparedCorpus[i].split(' ')
  for word in temp:
    c+=1
    if word not in my_count:
      my_count[word]=1
    else:
      my_count[word]+=1
  total_count.append(c)
  bag.append(my_count)
print(bag)
```

```
#Output bag of words
df = pd.DataFrame(bag)
df = df.fillna(0)
df = df.astype(int)
df
```

***Tf-Idf***

```
import copy
import math
import numpy as np
```

```
#Computing Tf
tf=copy.deepcopy(bag)
for i in range(len(tf)):
  for key,val in tf[i].items():
    tf[i][key]=val/total_count[i]
print(tf)
```

```
#'freq' tells in how many documents a particular word occurs
freq={}
for i in range(len(bag)):
  for key,value in bag[i].items():
    if key not in freq:
      freq[key]=1
    else:
      freq[key]+=1
freq
```

```
#Computing IDF
IDF = {}
for i in range(len(bag)):
  for key,val in bag[i].items():
    if key not in IDF:
      IDF[key]=math.log((1+len(bag))/(1+freq[key]))+1
print(IDF)
```

```
#Calculating tf-idf = tf*idf
tf_idf = []
for i in range(len(bag)):
  temp={}
  for key in IDF.keys():
    if key not in tf_idf:
      if key in tf[i]:
        temp[key]=tf[i][key]*IDF[key]
      else:
        temp[key]=0
  tf_idf.append(temp)
print(tf_idf)
```

```
df2 = pd.DataFrame(tf_idf)
df2 = df2.fillna(0)
df2
```

```
#Normalizing the products using root mean square
normal=[]
for i in range(len(bag)):
  x = df2.iloc[i,:].values
  x = np.sum(x**2)
  x = math.sqrt(x)
  normal.append(x)
normal
```

```
#final output 
for i in range(len(bag)):
  x = df2.iloc[i,:].values
  x=x/normal[i]
  df2.iloc[i,:]=x
df2
```

```
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
# df = pd.DataFrame(iris)
iris
```

```
independentVariables = pd.DataFrame(iris.data, columns = iris.feature_names)
dependentVariables = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```
independentVariables.info()
```

```
irisDataFrame = independentVariables.join(pd.Series(dependentVariables, name = 'class'))
irisDataFrame
```

```
between_class_feature_means = pd.DataFrame(columns = iris.target_names)
for c, rows in irisDataFrame.groupby('class'):
    between_class_feature_means[c] = rows.mean()
between_class_feature_means
```

```
within_class_scatter_matrix = np.zeros((4,4))
for c, rows in irisDataFrame.groupby('class'):
    rows = rows.drop(['class'], axis = 1)
    s = np.zeros((4,4))
for index, row in rows.iterrows():
    x, mc = row.values.reshape(4,1), between_class_feature_means[c].values.reshape(4,1)
    s += (x-mc).dot((x - mc).T)
within_class_scatter_matrix += s
feature_means = irisDataFrame.mean()
between_class_scatter_matrix = np.zeros((4,4))
for c in between_class_feature_means:
    n = len(irisDataFrame.loc[irisDataFrame['class'] == c].index)
    mc, m = between_class_feature_means[c].values.reshape(4,1), feature_means.values.reshape(4,1)
    between_class_scatter_matrix += n * (mc - m).dot((mc - m).T)
between_class_scatter_matrix
```

```
eigen_values, eigen_vectors = np.linalg.eig(np.linalg.inv(within_class_scatter_matrix).dot(between_class_scatter_matrix))
pairs = [(np.abs(eigen_values[i]), eigen_vectors[:,i]) for i in range (len(eigen_values))]
pairs = sorted(pairs, key = lambda x: x[0], reverse = True)
for pair in pairs:
    print(pair)
```

```
eigen_values_sum = sum(eigen_values)
print('Explained Variance: ')
for i, pair in enumerate(pairs):
    print('EigenVector: {}: {}'.format(i, (pair[0]/eigen_values_sum).real))
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X = np.insert(X,0,values=1,axis=1)
X = pd.DataFrame(X)
X.head()
```

```
k_fold = 5
group_size = X.shape[0] // k_fold
indexes = list(np.arange(0, group_size * k_fold))
input_grps = [[], [], [], [], []]
output_grps = [[], [], [], [], []]
for i in range(k_fold):
    lis = r.sample(indexes, group_size)
    for x in lis:
        #indexes.remove(x)
        input_grps[i].append(X.iloc[x])
        output_grps[i].append(y.iloc[x])
    input_grps[i] = pd.DataFrame(input_grps[i])
    output_grps[i] = pd.DataFrame(output_grps[i])
```

```
def find_Beta(X_train, y_train):
    A = X_train.T.dot(X_train)
    B=np.linalg.inv(A)
    C=B.dot(X_train.T)
    beta=C.dot(y_train)
    return(beta)
def find_predicted_values(X_test, beta):
    return X_test.dot(beta)
def find_RSquare(y_test, y_predict):
    sum_square_error=np.sum((y_test-y_predict)**2)
    mean_square_error=sum_square_error/len(y_predict)
    rms_error=np.sqrt(mean_square_error)
    y_mean=np.mean(y_test)
    total_variance=np.sum((y_test-y_mean)**2)
    return float(1-sum_square_error/total_variance)
def append_df(i,X_train,Y_train):
    for j in range(0,i):
        X_train = X_train.append(input_grps[j])
        Y_train = Y_train.append(output_grps[j])
    for j in range(i+1,len(input_grps)):
        X_train = X_train.append(input_grps[j])
        Y_train = Y_train.append(output_grps[j])
    return X_train, Y_train
    
Beta_matrices = []
Pred_matrices = []
R2_matrices = []
for i in range(k_fold):
    X_testing = input_grps[i]
    y_testing = output_grps[i]
    X_train = pd.DataFrame()
    Y_train = pd.DataFrame()
    X_training, y_training = append_df(i,X_train,Y_train)
    beta = find_Beta(X_training, y_training)
    Beta_matrices.append(beta)
    y_pred = find_predicted_values(X_testing, beta)
    Pred_matrices.append(y_pred)
    R2_matrices.append(find_RSquare(y_testing, y_pred))
```

```
print("BETA MATRICES:::\n",Beta_matrices)
print("PREDICTED VALUES:::\n",Pred_matrices)
print("R Square VALUES:::\n",R2_matrices)
```

```
#predict using best Beta from k_fold
from sklearn.model_selection import train_test_split
max_R2_index = R2_matrices.index(max(R2_matrices))
best_Beta = Beta_matrices[max_R2_index]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
y_pred_final = X_test.dot(best_Beta)
print("Predicted Values: \n",y_pred_final,"\n")
```

```
#Test error
error = y_test.values - y_pred_final.values.flatten()
sum_square_error = np.sum(error**2)
y_mean = np.mean(y_test)
total_variance = np.sum((y_test - y_mean)**2)
print("R2 Squared value is::\n",1 - (sum_square_error/total_variance))
```

```
#Implement Ridge regression using gradient descent optimization. Choose best parameters for which R2 is max and cost fn is min
x = np.array([i*np.pi/180 for i in range(60,300,4)])
np.random.seed(10) #Setting seed for reproducibility
df0 = pd.DataFrame(x, columns = ['x'])
for i in range(2,8): #power of 1 is already there
    colname = 'x_%d'%i #new var will be x_power
    df0[colname] = df0['x']**i
y = np.sin(x) + np.random.normal(0,0.15,len(x))
df0['y'] = y
df0.head()
```

```
X = df0.iloc[:,0:7]
y = df0.iloc[:,7]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X = pd.DataFrame(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 110)
X.head()
```

```
number_of_iterations = 50
n = X_train.shape[0]
regression_coefficients_forLR = []
for alpha in learning_rate:
    for lamba in regularization_parameter:
        beta = np.zeros(8)
        for i in range(number_of_iterations):
            x0_gradient = 0
            x1_gradient = 0
            x2_gradient = 0
            x3_gradient = 0
            x4_gradient = 0
            x5_gradient = 0    
            x6_gradient = 0
            x7_gradient = 0
            for j in range(len(X_train)):
                a = X_train.iloc[j,0]
                b = X_train.iloc[j,1]
                c = X_train.iloc[j,2]
                d = X_train.iloc[j,3]
                e = X_train.iloc[j,4] 
                f = X_train.iloc[j,5]
                g = X_train.iloc[j,6]
                h = y_train.iloc[j]
                x0_gradient += (beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)+(beta[6]*f)+(beta[7]*g)-h)
                x1_gradient += ((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)+(beta[6]*f)+(beta[7]*g)-h)*a)
                x2_gradient += ((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)+(beta[6]*f)+(beta[7]*g)-h)*b)
                x3_gradient += ((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)+(beta[6]*f)+(beta[7]*g)-h)*c)
                x4_gradient += ((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)+(beta[6]*f)+(beta[7]*g)-h)*d)
                x5_gradient += ((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)+(beta[6]*f)+(beta[7]*g)-h)*e)
                x6_gradient += ((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)+(beta[6]*f)+(beta[7]*g)-h)*f)
                x7_gradient += ((beta[0]+(beta[1]*a)+(beta[2]*b)+(beta[3]*c)+(beta[4]*d)+(beta[5]*e)+(beta[6]*f)+(beta[7]*g)-h)*g)
            val = alpha * lamba / n
            beta[0] = (beta[0] * (1 - val))-alpha/n*x0_gradient
            beta[1] = (beta[1] * (1 - val))-alpha/n*x1_gradient
            beta[2] = (beta[2] * (1 - val))-alpha/n*x2_gradient
            beta[3] = (beta[3] * (1 - val))-alpha/n*x3_gradient
            beta[4] = (beta[4] * (1 - val))-alpha/n*x4_gradient
            beta[5] = (beta[5] * (1 - val))-alpha/n*x5_gradient
            beta[6] = (beta[6] * (1 - val))-alpha/n*x6_gradient
            beta[7] = (beta[7] * (1 - val))-alpha/n*x7_gradient
        regression_coefficients_forLR.append(beta)
```

```
R2 = []
for i in range(len(regression_coefficients_forLR)):
    coeff = regression_coefficients_forLR[i]
    #R2 value calculation
    y_pred = X_test.dot(coeff[1:])
    y_pred += coeff[0]
    error = list(y_test - y_pred)
    square_error = np.power(error, 2)
    sum_square_error = 0
    for x in square_error:
        sum_square_error += float(x)
    y_mean = np.mean(y_test)
    total_variance = np.sum((y_test - y_mean)**2)
    R2.append(1-(sum_square_error/total_variance))
print(R2)
```

```
print("Best R2 value is: ",max(R2))
print("Best coefficients are: ",regression_coefficients_forLR[R2.index(max(R2))])
```

### **(i) Step by Step Implementation of CART**

```
class Node:
    def __init__(self, predicted_class):
        self.predicted_class = predicted_class
        self.feature_index = 0
        self.threshold = 0
        self.left = None
        self.right = None
```

```
class DecisionTreeClassifier:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth

    def fit(self, X, y):
        self.n_classes_ = len(set(y))
        self.n_features_ = X.shape[1]
        self.tree_ = self._grow_tree(X, y)

    def predict(self, X):
        return [self._predict(inputs) for inputs in X]

    def _best_split(self, X, y):
        m = y.size
        if m <= 1:
            return None, None
        num_parent = [np.sum(y == c) for c in range(self.n_classes_)]
        best_gini = 1.0 - sum((n / m) ** 2 for n in num_parent)
        best_idx, best_thr = None, None
        for idx in range(self.n_features_):
            thresholds, classes = zip(*sorted(zip(X[:, idx], y)))
            num_left = [0] * self.n_classes_
            num_right = num_parent.copy()
            for i in range(1, m):
                c = classes[i - 1]
                num_left[c] += 1
                num_right[c] -= 1
                gini_left = 1.0 - sum(
                    (num_left[x] / i) ** 2 for x in range(self.n_classes_)
                )
                gini_right = 1.0 - sum(
                    (num_right[x] / (m - i)) ** 2 for x in range(self.n_classes_)
                )
                gini = (i * gini_left + (m - i) * gini_right) / m
                if thresholds[i] == thresholds[i - 1]:
                    continue
                if gini < best_gini:
                    best_gini = gini
                    best_idx = idx
                    best_thr = (thresholds[i] + thresholds[i - 1]) / 2
        return best_idx, best_thr

    def _grow_tree(self, X, y, depth=0):
        num_samples_per_class = [np.sum(y == i) for i in range(self.n_classes_)]
        predicted_class = np.argmax(num_samples_per_class)
        node = Node(predicted_class=predicted_class)
        if depth < self.max_depth:
            idx, thr = self._best_split(X, y)
            if idx is not None:
                indices_left = X[:, idx] < thr
                X_left, y_left = X[indices_left], y[indices_left]
                X_right, y_right = X[~indices_left], y[~indices_left]
                node.feature_index = idx
                node.threshold = thr
                node.left = self._grow_tree(X_left, y_left, depth + 1)
                node.right = self._grow_tree(X_right, y_right, depth + 1)
        return node

    def _predict(self, inputs):
        node = self.tree_
        while node.left:
            if inputs[node.feature_index] < node.threshold:
                node = node.left
            else:
                node = node.right
        return node.predicted_class
```

```
clf = DecisionTreeClassifier(max_depth = 100)
clf.fit(X_train.to_numpy(), y_train.to_numpy())
y_pred = clf.predict(X_test.to_numpy())
```
